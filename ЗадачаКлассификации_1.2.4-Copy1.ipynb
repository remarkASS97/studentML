# Задание 1.2
- создать функцию summ_pok которая будет подсчитывать сумму по данным столбцам:
- - HS + AS = HAS
- - HST + AST = HAST
- - HC + AC = HAC
- - HF + AF = HAF
- - HY + AY = HAY
- - HR + AR = HAR

```python
def summ_pok(df: pd.DataFrame) -> pd.DataFrame:
    """Function to calculate sum by columns
    :param df: Original DataFrame
    :return: sum by columns
    """
    result_df = pd.DataFrame()
    result_df['HAS'], result_df['HAST'], result_df['HAC'], result_df['HAF'], result_df['HAY'], result_df['HAR'] = \
    df.HS + df.AS, df.HST + df.AST, df.HC + df.AC, df.HF + df.AF, df.HY + df.AY, df.HR + df.AR
    return result_df
```

```python
def pivot(df: pd.DataFrame, index: str):
    """Function to create pivot table by 'index' column
    :param df: Original DataFrame
    :index: column on which the pivot table is created
    :return: pivot table
    """
    table = pd.pivot_table(df, index=index, aggfunc=np.sum)
    if index == 'H':
        del table['A']
    elif index == 'A':
        del table['H']
    return(table.reset_index())
```

```python
def pivot_S(df: pd.DataFrame, index: str):
    """Function to create pivot table by 'index' column for all S
    :param df: Original DataFrame
    :index: column on which the pivot table is created
    :return: pivot table for all S
    """
    table = pd.pivot_table(df, index=['S', index], aggfunc=np.sum)
    if index == 'H':
        del table['A']
    elif index == 'A':
        del table['H']
    return(table.reset_index())
```

```python
def cumsum_os_ha(data: pd.DataFrame):
    data['p1'] = data.groupby(['S','H'])['ph'].transform(lambda x: x.cumsum().shift())
    data['p2'] = data.groupby(['S','A'])['pa'].transform(lambda x: x.cumsum().shift())
    data['p5'] = data.groupby(['H'])['ph'].transform(lambda x: x.cumsum().shift())
    data['p6'] = data.groupby(['A'])['pa'].transform(lambda x: x.cumsum().shift())
    return data
```

```python
def cumsum_o_ha(data: pd.DataFrame):
    data['id'] = range(1, len(data) + 1)
    data1 = data[['S', 'H', 'ph', 'id']]
    data2 = data[['S', 'A', 'pa', 'id']]
    data2.rename(columns={'A':'H', 'pa': 'ph'}, inplace=True)
    data3 = pd.concat([data1, data2])
    data3.sort_values('id', inplace=True)
    data3['p3'] = data3.groupby(['S','H'])['ph'].transform(lambda x: x.cumsum().shift())
    data = data.merge(data3, left_on=('S', 'H', 'id', 'ph'), right_on=('S', 'H', 'id', 'ph'))
    data3.rename(columns={'H':'A', 'ph': 'pa'}, inplace=True)
    data = data.merge(data3, left_on=('S', 'A', 'id', 'pa'), right_on=('S', 'A', 'id', 'pa'))
    data.rename(columns={'p3_x': 'p3', 'p3_y': 'p4'}, inplace=True)
    
    return data
```

```python
def cumsum_o(data: pd.DataFrame):
    data['id'] = range(1, len(data) + 1)
    data1 = data[['H', 'ph', 'id']]
    data2 = data[['A', 'pa', 'id']]
    data2.rename(columns={'A':'H', 'pa': 'ph'}, inplace=True)
    data3 = pd.concat([data1, data2])
    data3.sort_values('id', inplace=True)
    data3['p7'] = data3.groupby(['H'])['ph'].transform(lambda x: x.cumsum().shift())
    data = data.merge(data3, left_on=('H', 'id', 'ph'), right_on=('H', 'id', 'ph'))
    data3.rename(columns={'H':'A', 'ph': 'pa'}, inplace=True)
    data = data.merge(data3, left_on=('A', 'id', 'pa'), right_on=('A', 'id', 'pa'))
    data.rename(columns={'p7_x': 'p7', 'p7_y': 'p8'}, inplace=True)
    return data
```
