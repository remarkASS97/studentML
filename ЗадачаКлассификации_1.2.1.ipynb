# Задание 1.2
- создать функцию summ_pok которая будет подсчитывать сумму по данным столбцам:
- - HS + AS = HAS
- - HST + AST = HAST
- - HC + AC = HAC
- - HF + AF = HAF
- - HY + AY = HAY
- - HR + AR = HAR

```python
def summ_pok(df: pd.DataFrame) -> pd.DataFrame:
    """Function to calculate sum by columns
    :param df: Original DataFrame
    :return: sum by columns
    """
    result_df = pd.DataFrame()
    result_df['HAS'], result_df['HAST'], result_df['HAC'], result_df['HAF'], result_df['HAY'], result_df['HAR'] = \
    df.HS + df.AS, df.HST + df.AST, df.HC + df.AC, df.HF + df.AF, df.HY + df.AY, df.HR + df.AR
    return result_df
```

```python
summ_pok(df)
```

```python
new_table = pd.concat([df, summ_pok(df)], axis=1)
new_table
```

# Задание 1.3
- создать сводную таблицу по столбцу H (аналогично для столбца A)
- - взять все уникальные объекты в колонке H
- - сводную информацию за все S по столбцам gh	ga	allg	ph	pa	HS	AS	HST	AST	HC	AC	HF	AF	HY	AY	HR	AR y1 y2(суммировать)

```python
def pivot(df: pd.DataFrame, index: str):
    """Function to create pivot table by 'index' column
    :param df: Original DataFrame
    :index: column on which the pivot table is created
    :return: pivot table
    """
    table = pd.pivot_table(df, index=index, aggfunc=np.sum)
    if index == 'H':
        del table['A']
    elif index == 'A':
        del table['H']
    return(table.reset_index())
```

```python
pivot(df, index = 'H')
```

```python
pivot(df, index = 'A')
```

```python
def pivot_S(df: pd.DataFrame, index: str):
    """Function to create pivot table by 'index' column for all S
    :param df: Original DataFrame
    :index: column on which the pivot table is created
    :return: pivot table for all S
    """
    table = pd.pivot_table(df, index=['S', index], aggfunc=np.sum)
    if index == 'H':
        del table['A']
    elif index == 'A':
        del table['H']
    return(table.reset_index())
```

```python
pivot_S(df, 'H')
```

```python
pivot_S(df, 'A')
```
